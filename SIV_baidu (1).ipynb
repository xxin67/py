```python
#读取秘钥csv文件中的key

import csv
# 储存着key的文件
key_path = r'key.csv'
# 创建一个list 记录所有的key
key_list = []
with open (key_path, mode = 'r', encoding = 'utf-8') as f:
    lines = csv.reader(f)
    for line in lines:
        print(line)
        key_list.append(line)
key_list    
```

```python
# 读取秘钥csv文件中的序号和坐标

# 储存着坐标的文件
csv_path = r'zb_excle.csv'
# 创建一个list记录所有的 key
data_list = []
with open(csv_path, mode = 'r', encoding = 'utf-8')as f:
    lines = csv.reader(f)
    for line in lines:
        data_list.append(line)
data_list = data_list[1:] # 把第一行删去
data_list
```

```python
# 坐标转换
import requests
import json
import time

def wgs2bd09mc(lon, lat):
    url = 'https://api.map.baidu.com/geoconv/v2/?coords={},{}&from=1&to=5&ak=HZU0Pbugrb5MlRZEUxIVBn0jGUwJa4dG'.format(lon,lat)

    response = requests.get(url)
    response = response.content
    response = response.decode()
    temp = json.loads(response)


    if temp ['status'] == 0:
        bd09_lon = temp['result'][0]['x']
        bd09_lat = temp['result'][0]['y']
    return bd09_lon, bd09_lat
```

```python
# 使用lon,lat,lat下载街景
# 循环拿到FID，lon，lat，key
for data in data_list:
    print(data)
    FID,lon,lat = data[0],data[1],data[2]
    lon,lat = wgs2bd09mc(lon,lat)
    time.sleep(0.5)
    print(lon,lat)
```
